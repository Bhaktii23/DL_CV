```python
import cv2
import os
# from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
# Load image
img = plt.imread('/content/flower.jpg')
```

```python
# Display original image
# cv2_imshow(img)
plt.imshow(img)
# cv2.waitKey(0)
```

```python
# Resizing image
img_resize = cv2.resize(img,(300,300))
plt.imshow(img_resize)
# cv2_imshow(img_resize)
cv2.waitKey(0)
```

```python
# Cropping Image
img_crop=img[100:300,200:500]
plt.imshow(img_crop)
# cv2_imshow(img_crop)
cv2.waitKey(0)
```

```python
# Apply binary threshold
thresholded_img = cv2.threshold(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY), 125, 255,
cv2.THRESH_BINARY)
# cv2_imshow(thresholded_img[1])
plt.imshow(thresholded_img[1])
cv2.waitKey(0)
```

```python
#contour analysis
contours,_ = cv2.findContours(thresholded_img[1], cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)  # Use thresholded_img[1] to access the thresholded image
contour_img = img.copy()
cv2.drawContours(contour_img, contours, -4, (0,255,0), 7)
plt.imshow(contour_img);
```

```python
# Blob detection (SimpleBlobDetector)
import numpy as np
detector = cv2.SimpleBlobDetector_create()
keypoints = detector.detect(img)
blob_img = img.copy()
cv2.drawKeypoints(blob_img, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(blob_img);
```
